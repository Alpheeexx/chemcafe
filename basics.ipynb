## Decisions, Decisions


### Motivating Example:


This has almost no practical use, but let's make a script to determine how old someone was 7 years ago! Our first attempt might be something like this.


```python
name = input("Enter the name of someone: ")
age = int(input("Enter their age: "))

old_age = age - 7

print("Seven years ago, " + name + " was " + str(old_age) + " years old!")
```

But, we have a bit of a problem. What if the person you enter is younger than 7? Seeing **-5 years old** just doesn't make sense. Instead, we'll use an _if statement_.


```python
name = input("Enter the name of someone: ")
age = int(input("Enter their age: "))

old_age = age - 7

if old_age < 0:
    print("Seven years ago, " + name + " wasn't even born!")
else:
    print("Seven years ago, " + name + " was " + str(old_age) + " years old!")
```

You might notice we have a bit of repeated text here. We can clean it up a bit by storing the first part of the string in a variable like this:


```python
name = input("Enter the name of someone: ")
age = int(input("Enter their age: "))

start = "Seven years ago, " + name + " "

old_age = age - 7

if old_age < 0:
    print(start + "wasn't even born!")
else:
    print(start + "was " + str(old_age) + " years old!")
```

Again, you'll notice that if someone who's exactly 7 years old has their details entered, we get **0 years old**, which doesn't sound quite right. And if someone's 8 years old, we get **1 years old**, which doesn't really follow the rules of English grammar. Let's fix one of those problems:


```python
name = input("Enter the name of someone: ")
age = int(input("Enter their age: "))

start = "Seven years ago, " + name + " "

old_age = age - 7

if old_age < 0:
    print(start + "wasn't even born!")
elif old_age == 0:
    print(start + "was less than a year old!")
else:
    print(start + "was " + str(old_age) + " years old!")
```

## Repetition, Repetition


### Example:

Now, I'm not sure if you noticed something about a few of the later exercises last week, but you had to repeat yourself a lot. Let's pick an even more tedious example! This script takes as input 7 numbers, and then outputs their sum.


```python
num1 = int(input("Number 1 please "))
num2 = int(input("Number 2 please "))
num3 = int(input("Number 3 please "))
num4 = int(input("Number 4 please "))
num5 = int(input("Number 5 please "))
num6 = int(input("Number 6 please "))
num7 = int(input("Number 7 please "))

total = num1 + num2 + num3 + num4 + num5 + num6 + num7
print(total)
```

This code has two major 'smells' - the first is that we don't actually need this many variables, as much as this script is structured like the problems from last week. We can immediately discard the input values after adding them to the total, like so:


```python
total = 0
total += int(input("Number 1 please "))
total += int(input("Number 2 please "))
total += int(input("Number 3 please "))
total += int(input("Number 4 please "))
total += int(input("Number 5 please "))
total += int(input("Number 6 please "))
total += int(input("Number 7 please "))

print(total)
```

The second code smell is the fact that we're basically doing the same thing 7 times, with a minor difference in the string used as prompt for the input each time. This seems like a good opportunity to introduce `for` loops!


```python
total = 0
for i in range(7):
    text = input("Number " + str(i) + " please ")
    total += int(text)

print(total)
```

So, what happened there? You can effectively think of a for loop as something that will run a section of code over and over again, changing the value of (in this case) `i` each time to go through everything inside the function `range(7)`. In this case, it's all the numbers from 0 up to 6. After all of these iterations, we move past the loop. Pretty neat huh?

But, that isn't exactly the same output as our first version of the program.
**Can you fix it?**

## Lists

There is another way we can make things nicer for ourselves - lists! They're basically just a way for us to store related chunks of data.


```python
items = ["Apples", "Oranges", "Bananas"]
stock = [6, 4, 8]

# You can add to them!
items.append("Grapes")
stock.append(2)

# You can change (zero-indexed) elements of them!
items[1] = "Tangerines"
stock[0] -= 3

# You can also remove from them!
items.remove("Bananas")
stock.remove(4)

# We can use loops to iterate through them
print("My Stock: ")
for i in range(len(items)):  # Using the length of the items list
    print("I have " + str(stock[i]) + " " + items[i])
```

Try playing around with the script above, especially if you're unsure of a few of the concepts.

We can also iterate over arrays, like so


```python
for item in items:
    print(item)
```

Also, surprise! You've actually been dealing with something list-like already, as a string is basically just a list of characters. However, strings are immutable - meaning we can't just randomly replace characters in them.


```python
s = "Duck"
s[0] = "F"
```

We can still iterate over them with for loops, however - as we'll come to find.
