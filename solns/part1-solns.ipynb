# Manipulating Variables

## Exercise 1.1:

```python
hello_friend = 5
hello_friend += 10
print(hello_friend)
```

When you first run the cell, it should output 15. When you swap the first two lines and run it again, you should get an error that `hello_friend` is not defined. So `+=` must add to an existing variable. As such, the following two statements are equivalent:

```python
# Using +=
a = 7
a += 42

# Without +=
b = 7
b = b + 42

print(a, b)
```

## Exercise 1.2:

The `*` operator is multiplication.

The `//` operator is integer (floor) division, so will round down to the next integer. This tripped a few people up, who mistook it for a regular division.

Note that the `/` division operator always returns a `float`, not an `int` - even if it divides without remainder. See below:

```python
print(12 * 5)

print(12 / 5)
print(12 // 5)

print(12 / 4)
print(12 // 4)
```

The only modification needed is setting `a = 30`. This is because we subtract `b` from it on the third line, which is also 30. Finally, we try to divide by 0 on the last line, which causes an error. Try running the script below:

```python
a = 30
b = 6 * 5
a -= b
print(100 // a)
```

## Exercise 1.4: 

Firstly, sorry about not saying in the problem description that you needed to use `math.sqrt`! Other than that, the calculations should be reasonably straight-forward after copying the example for a square.

```python
import math

width = int(input("Input Width: "))
height = int(input("Input Height: "))

area = width * height / 2
hypotenuse = math.sqrt(width * width + height * height)
perimeter = width + height + hypotenuse

print("Area: " + str(area) + " Perimeter: " + str(perimeter))
```

## Exercise 1.5:

Surprise! You only need the area or perimeter, not both, to determine the radius of a semicircle. So you can choose to ignore one of the inputs entirely. Let's use the area for this solution. Again, we'll need `math.sqrt`, but also `math.pi` this time.

```python
import math

area = int(input("Area: "))
_ = int(input("Perimeter: "))

r_squared = (area * 2) / math.pi
r = math.sqrt(r_squared)

print("Radius: " + str(r))
```

# Decisions, Decisions

## Motivating Example:

The last part of the example asked you to modify the script such that if someone enters the age as being 8, then we correct **1 years old** to make grammatical sense. Here's one way to do that:

```python
name = input("Enter the name of someone: ")
age = int(input("Enter their age: "))

start = "Seven years ago, " + name + " "

old_age = age - 7

if old_age < 0:
  print(start + "wasn't even born!")
elif old_age == 0:
  print(start + "was less than a year old!")
elif old_age == 1:
  print(start + "was a year old!")
else:
  print(start + "was " + str(old_age) + " years old!")
```

Here's another way, using Python f-strings (look them up!) to implicitly handle datatype conversion:

```python
name = input("Name: ")
age = int(input("Age: "))

start = f"Seven years ago, {name}"

if age < 7:
  print(f"{start} wasn't even born!")
elif age == 7:
  print(f"{start} was less than a year old!")
else:
  plural = "s"
  if age == 8:
    plural = ""
  print(f"{start} was {age - 7} year{plural} old!")
```

## Exercise 2.2:

We can split the problem into the cases where x, y and z are the hypotenuse. Make sure to check that the lengths are non-negative first! Here I've used the max function, but you can just tpye `x < z and y < z` to replace `max(x, y) < z`.

```python
x = int(input("x: "))
y = int(input("y: "))
z = int(input("z: "))

if x <= 0 or y <= 0 or z <= 0:
  print("No")
elif max(x, y) < z and x * x + y * y == z * z:
  print("Yes")
elif max(x, z) < y and x * x + z * z == y * y:
  print("Yes")
elif y * y + z * z == x * x: # x must be the largest
  print("Yes")
else:
  print("No")
```

## Exercise 2.3:

Quadratic formula time! But watch out for if there's one solution or no solutions.

```python
import math

a = int(input("a: "))
b = int(input("b: "))
c = int(input("c: "))

# Equation is the following: (-b +- sqrt(b * b - 4 * a * c )) / (2 * a)

discrim = b * b - 4 * a * c
if discrim < 0:
  print("No Solutions")
elif discrim == 0:  # 1 solution
  print(- b / (2 * a))
else:  # 2 solutions
  s = math.sqrt(discrim)
  print(- b + s / (2 * a))
  print(- b - s / (2 * a))
```
